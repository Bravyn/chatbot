### SmartDataframe

A SmartDataframe is a pandas (or polars) dataframe that inherits all the properties and methods from the `pd.DataFrame`, but also adds conversational features to it.

```python
from pandasai import SmartDataframe
import pandas as pd
import os
```

```python
df = pd.read_csv("/content/Most Streamed Spotify Songs 2024.csv", encoding='unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
os.environ['PANDASAI_API_KEY'] = ""
```

```python
df.columns
```

```python
df.info()
```

```python
float_vars = ['All Time Rank', 'Spotify Streams',
       'Spotify Playlist Count', 'Spotify Playlist Reach',
      'YouTube Views', 'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'TikTok Views', 'YouTube Playlist Reach',
       'AirPlay Spins', 'SiriusXM Spins',
       'Deezer Playlist Count', 'Deezer Playlist Reach',
       'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts', 'TIDAL Popularity']

def remove_commas(x):
    return x.replace(',', '')

for var in float_vars:
    df[var] = df[var].apply(remove_commas).astype(float)

df.info()
```

```python
sdf = SmartDataframe(df)
```

```python
sdf.chat("Return top 5 rows")
```
